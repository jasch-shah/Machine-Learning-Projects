```python
# most ordered item
c = chip.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

```python
# most ordered item in choice_description col
c = chip.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending = False)
c.head(1)
```

```python
# items were ordered in total
total_items_orders = chip.quantity.sum()
total_items_orders
```

```python
# price into float
chip.item_price.dtype
```

```python
dollarizer = lambda x: float(x[1:-1])
chip.item_price = chip.item_price.apply(dollarizer)
```

```python
chip.item_price.dtype
```

```python
# how much was revenue for period in dataset
revenue = (chip['quantity']* chip['item_price']).sum()
print('Revenue was $ '+ str(np.round(revenue, 2)))
```

```python
orders = chip.order_id.value_counts().count()
orders
```

```python
# Average revenue amount per order 
# Solution 1
chip['revenue'] = chip['quantity'] * chip['item_price']
order_grouped = chip.groupby(by=['order_id']).sum()
order_grouped.mean()['revenue']
```

```python
# Solution 2
chip.groupby(by=['order_id']).sum().mean()['revenue']
```
